```python
# Import the dependencies
import pandas as pd
```

```python
# source dataset: https://hdr.undp.org/data-center/human-development-index#/indicies/HDI
# Set the column width
pd.options.display.max_colwidth = 200
# Read in the dataset. 
hdiDF = pd.read_excel('./HDI_Table.xlsx')
hdiDF
```

```python
# Get the index datatype, column names.
hdiDF.info()
```

```python
# source dataset: https://www.kaggle.com/datasets/ashishraut64/internet-users
# read the CSV file into a Pandas DataFrame
df = pd.read_csv('./Int User.csv')

# Select only the rows where the "Year" column equals 2020
df_2020 = df[df["Year"] == 2020]

# Drop any rows where the "Entity" column is blank or null
df_2020 = df_2020.dropna(subset=["Entity"])

# Reset the index of the new DataFrame
df_2020 = df_2020.reset_index(drop=True)
```

```python
# Select only the rows where the "Year" column equals 2020
df_2020.head()
```

```python
# drop column 'Unnamed:0 and 'Year'
df_2020 = df_2020.drop(['Unnamed: 0', 'Year'], axis=1)
```

```python
# rename column 'Entity' to 'Country'
df_2020 = df_2020.rename(columns={'Entity': 'Country'})
# print dataframe
df_2020.head()
```

```python
# merge 2 dataframes on 'Country' column
merged_df = pd.merge(hdiDF, df_2020, on='Country')

merged_df
```

```python
# Drop column 'HDI rank(2020)'
merged_df = merged_df.drop(['HDI rank(2020)'], axis=1)
```

```python
# check data type in the merged dataframe
print(merged_df.dtypes)
```

```python
# export dataframe to excel
merged_df.to_excel('merged_data.xlsx', index=False)
```

```python
# list all column name in merged_df dataframe
columns = merged_df.columns

print(columns)
```

```python
# check to make sure that there is no duplicated column name.
print(merged_df.columns.duplicated())
```

```python
# Assigned column order
column_order = ['HDI rank', 'Country', 'Code', 
        'Human Development Index (HDI) 2021',
       'Life expectancy at birth (2021)', 'Expected years of schooling (2021)',
       'Mean years of schooling (2021)',
       'Gross national income (GNI) per capita (2021)',
       'GNI per capita rank minus HDI rank (2021)',  
       'Cellular Subscription', 'Internet Users(%)', 
       'No. of Internet Users',
       'Broadband Subscription']

# Reorder the columns in the merged_df DataFrame
merged_df = merged_df.reindex(columns=column_order)

merged_df.head()
```

## LOAD INTO MONGODB

```python
from pprint import pprint
```

```python
# Convert the dataframe to a list of dictionaries
data = merged_df.to_dict('records')
#data
```

```python
from pymongo import MongoClient

# Create an instance of MongoClient
mongo = MongoClient(port=27017)

# Select the database and collection where you want to insert the data
db = mongo.hdiDB
hdiInfo = db.hdiInfo
```

```python
# Insert the data into the MongoDB collection
hdiInfo.insert_many(data)
```

```python
# query the collection
results = list(hdiInfo.find())
```

```python
# checked to see if the data was loaded
for x in results:
    pprint(x)
```
